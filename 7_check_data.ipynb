```python
import ftfy
# Rồi fix lỗi từng cột
for col in ['Film', 'User', 'Review', 'EmotionLabel']:
    tmp_df[col] = tmp_df[col].fillna('').apply(ftfy.fix_text)
```

```python
from ftfy import fix_text
from unidecode import unidecode

def clean_text(text):
    if isinstance(text, str):
        fixed = fix_text(text)
        return unidecode(fixed)
    return text

for col in ['Film', 'User', 'Review', 'EmotionLabel']:
    tmp_df[col] = tmp_df[col].apply(clean_text)
```
